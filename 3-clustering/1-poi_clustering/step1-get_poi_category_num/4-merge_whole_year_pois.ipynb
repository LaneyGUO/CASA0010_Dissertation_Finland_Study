# 1 Import package

```python
import pandas as pd
import os
```

# 2 Merge data

```python
# Input and output folder paths
input_folder = '../../../../data/Locomizer_edited/pois/add_category_number'
output_folder = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT_NUMBER'

# Create output folder
os.makedirs(output_folder, exist_ok=True)
```

```python
# Initialize an empty list to store the merged DataFrame
merged_df_list = []

# Iterate over the CSV files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.csv'):
        # Construct the input file path
        input_csv_file = os.path.join(input_folder, filename)
        
        # Read the CSV file
        df = pd.read_csv(input_csv_file)
        
        # Remove rows where the CAT_NUMBER column is empty
        df = df[df['CAT_NUMBER'].notna()]
        
        # Add the current DataFrame to the list
        merged_df_list.append(df)

# Merge all DataFrames
merged_df = pd.concat(merged_df_list, ignore_index=True)
```

# 3 Check Duplicated rows

```python
# Check for duplicate lines
duplicate_rows = merged_df[merged_df.duplicated()]
duplicate_count = duplicate_rows.shape[0]
```

```python
# There are 1845342 duplicated rows. Deleted them
distinct_df = merged_df.drop_duplicates()

distinct_df.head()
```

# 4 Export dataframe

```python
# Save the merged DataFrame to a new CSV file
output_csv_file = os.path.join(output_folder, 'whole_year_poi_with_CAT_NUMBER.csv')
distinct_df.to_csv(output_csv_file, index=False)
```
