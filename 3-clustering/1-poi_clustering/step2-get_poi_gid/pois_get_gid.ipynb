# 1 Import package

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, shape
import os
import ijson
```

# 2 Join data

```python
# Read CSV file of POIs
poi_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT_NUMBER/whole_year_poi_with_CAT_NUMBER.csv'
df_poi = pd.read_csv(poi_file_path)

df_poi.head()
```

```python
df_poi.info()
```

```python
geojson_file_path = '../../../../data/Locomizer/finland_lvl10.geojson'
features = []

with open(geojson_file_path, 'r') as f:
    for item in ijson.items(f, 'features.item'):
        geom = shape(item['geometry'])  # Convert geometry to Shapely objects
        properties = item['properties']  # Get Properties
        features.append({'geometry': geom, **properties})

gdf = gpd.GeoDataFrame(features, crs='EPSG:4326')
```

```python
gdf.info()
```

```python
gdf.head(12)
```

```python
# Save gdf as a CSV file
output_file_path = '../../../../data/Locomizer_edited/finland_hexagons.csv'
gdf.to_csv(output_file_path, index=False)
```

```python
# Create GeoDataFrame of POIs
geometry = [Point(xy) for xy in zip(df_poi['LATITUDE'], df_poi['LONGITUDE'])]
crs = {'init': 'epsg:4326'}
gdf_poi = gpd.GeoDataFrame(df_poi, crs=crs, geometry=geometry)
```

```python
gdf_poi.head()
```

```python
gdf_poi.info()
```

```python
# Spatial join
joined = gpd.sjoin(gdf_poi, gdf, how='left', op='intersects')
```

```python
joined.info()
```

```python
joined.head()
```

```python
# Add gid_10 column to original POI DataFrame
df_poi['gid_10'] = joined['gid_10_right']
```

```python
df_poi.info()
```

```python
df_poi.head()
```

# 3 Export files

## 3.1 Export poi and gid data

```python
# Save df_poi as a CSV file
output_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT&gid.csv'
df_poi.to_csv(output_file_path, index=False)
```

## 3.2 Export poi data with spatial(polygon) information

```python
# Copy df_poi to df_poi_geo, make sure it's a deep copy
df_poi_geo = df_poi.copy(deep=True)

# Get the required columns by merge
df_poi_geo = df_poi_geo.merge(
    gdf[['gid_10', 'geometry', 'center_lat', 'center_lon']],
    on='gid_10',
    how='left'
)

# Rename merged columns
df_poi_geo.rename(columns={
    'geometry': 'polygon_geometry',
    'center_lat': 'center_lat',
    'center_lon': 'center_lon'
}, inplace=True)
```

```python
# Export df_poi_geo as a CSV file
output_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT&gid&geometry.csv'
df_poi_geo.to_csv(output_file_path, index=False)
```
