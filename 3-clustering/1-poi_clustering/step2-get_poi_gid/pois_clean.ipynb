POIs duplicates check again, removing the same rows and keeping only the unique rows.

# 1 Import package

```python
import pandas as pd
```

# 2 Check duplicated data

```python
# Read CSV file of POIs
input_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT&gid&geometry.csv'
df = pd.read_csv(input_file_path)

df.head()
```

```python
df.info()
```

```python
# Grouped and sorted by gid_10
sorted_df = df.sort_values(by=['gid_10', 'LATITUDE', 'LONGITUDE'])
sorted_df.head()
```

```python
# Remove duplicate rows, keeping only the first row in each group with the same LATITUDE, LONGITUDE, NAME
df_clean = sorted_df[~sorted_df.duplicated(subset=['gid_10', 'LATITUDE', 'LONGITUDE', 'NAME'], keep='first')]
df_clean.head()
```

```python
df_clean.info()
```

# 3 Export files

## 3.1 Export poi and gid data

```python
df_clean_without_geometry = df_clean.drop(['polygon_geometry', 'center_lat', 'center_lon'], axis=1)
```

```python
df_clean_without_geometry.info()
```

```python
# Save df_poi as a CSV file
output_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT&gid_clean.csv'
df_clean_without_geometry.to_csv(output_file_path, index=False)
```

## 3.2 Export poi data with spatial(polygon) information

```python
# Export df_poi_geo as a CSV file
output_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT&gid&geometry_clean.csv'
df_clean.to_csv(output_file_path, index=False)
```
