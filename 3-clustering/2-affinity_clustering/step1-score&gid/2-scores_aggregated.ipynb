# 1 Import package

```python
import pandas as pd
```

# 2 Aggregate data and calculate average scores for each gid_10 in 15 categories

```python
file_path = '../../../../data/Locomizer_edited/profiling/whole_year_scores_with_gid.txt'
df = pd.read_csv(file_path, sep='\t')
df.head()
```

```python
df.info()
```

```python
# Group the columns by ‘gid_10’ and calculate the average of columns 101 to 115 for each group
grouped_df = df.groupby('gid_10')[[str(i) for i in range(101, 116)]].mean()

grouped_df.info()
grouped_df.head()
```

```python
# Reset index ‘gid_10’ to a column
grouped_df = grouped_df.reset_index()

# Check if ‘gid_10’ column is unique
# "True"
print("Is 'gid_10' unique after reset index? ", grouped_df['gid_10'].is_unique)
```

# 3 Export data

```python
# Save the aggregated DataFrame to a text file
output_path = '../../../../data/Locomizer_edited/profiling/whole_year_scores_with_gid_average.txt'
grouped_df.to_csv(output_path, sep='\t', index=False)
```
