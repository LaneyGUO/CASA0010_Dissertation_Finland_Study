# 1 Import package

```python
import pandas as pd
```

# 2 Integrated processing data

```python
# Read CSV file of POIs
input_file_path = '../../../../data/Locomizer_edited/pois/whole_year_poi_with_CAT&gid&geometry_clean.csv'
df = pd.read_csv(input_file_path)
```

```python
df.info()
```

```python
# Create a DataFrame containing gid_10 and CAT_NUMBER
# Count the occurrences of CAT_NUMBER from 101 to 115
count_df = df.groupby('gid_10')['CAT_NUMBER'].value_counts().unstack(fill_value=0)
count_df.head(20)
```

```python
# # Checks if there is an odd number of columns starting from the second column and returns information about the rows with odd numbers
# odd_rows = count_df.loc[(count_df.iloc[:, 1:] % 2 != 0).any(axis=1)]

# # Get the number of rows with an odd number
# odd_rows_count = len(odd_rows)

# # output results
# print(f"Number of rows with odd numbers: {odd_rows_count}")
# print("Row information with odd numbers:\n", odd_rows)
```

```python
# Keep only the columns from 101 to 115
count_df = count_df.reindex(columns=range(101, 116), fill_value=0)

# Reset indexes to put gid_10 column back into DataFrame
count_df.reset_index(inplace=True)

count_df.head()
```

```python
# Merge the count results with the raw data to get polygon_geometry, centre_lat, centre_lon
# Select additional information corresponding to gid_10
additional_info = df[['gid_10', 'polygon_geometry', 'center_lat', 'center_lon']].drop_duplicates()
additional_info.head()
```

```python
# Merge data
final_df = pd.merge(count_df, additional_info, on='gid_10', how='left')
final_df.head()
```

```python
final_df.info()
```

# 3 Export files

```python
# Save hexagons with category volumns dataframe as a CSV file
output_file_path = '../../../../data/Locomizer_edited/pois/aggregated_poi_counts_clean.csv'
final_df.to_csv(output_file_path, index=False)
```
